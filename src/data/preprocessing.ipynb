```python
import pandas as pd
from xml.dom import minidom
import spacy
import os
from tqdm import tqdm
```

```python
PATH_DATA = '../../dataset/'
```

```python
mydoc = minidom.parse(PATH_DATA + 'raw/alter_full.xml')
predicate = mydoc.getElementsByTagName('predicate')[0].getAttribute('lemma')
roles = [role.getAttribute('descr') for role in mydoc.getElementsByTagName('role')]
examples = mydoc.getElementsByTagName('example')
ids = [i for i in range(len(examples))]
srcs, texts, args = [], [], []
print(roles)
for example in examples:
    text = example.getElementsByTagName('text')[0].firstChild.nodeValue
    src = example.getAttribute('src')
    # arg_temp = ["" for i in range(len(self.roles))]
    arg_temp = dict()
    for role in roles:
        arg_temp[role] = ""
        print(arg_temp)
    # print(arg_temp)
    for arg in example.getElementsByTagName('arg'):
    # arg_temp.append(arg.firstChild.nodeValue)
        arg_temp.update({arg.getAttribute('n'): arg.firstChild.nodeValue})
    texts.append(text)
    srcs.append(src)
    args.append(arg_temp)
data_arg = pd.DataFrame({'id': ids, 'source': srcs, 'text': texts, 'arguments': args})
```
